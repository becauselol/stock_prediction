```python
from neo4j import GraphDatabase, RoutingControl
```

```python
import pandas as pd
import neo4j, tqdm
```

```python
import json
```

```python
URI = "neo4j://localhost:11006"
AUTH = ("neo4j", "nedo")
```

```python
selected_tickers = pd.read_csv("../selected_tickers.csv")
```

```python
tickers = list(selected_tickers["ticker"])
```

```python
def my_query(driver, ticker):
    records, _, _ = driver.execute_query(
        """MATCH (n:Company) where n.ticker = \"""" + ticker + """\" CALL apoc.path.subgraphAll(n, {labelFilter: "-Fact", maxLevel:1}) YIELD nodes, relationships RETURN nodes, relationships""",
        database_="neo4j", routing_=RoutingControl.READ
    )
    return records
```

```python
with GraphDatabase.driver(URI, auth=AUTH) as driver:
        rec = my_query(driver, "AAPL")
```

```python
driver = GraphDatabase.driver(URI, auth=AUTH)
session = driver.session()
```

```python
def new_query(ticker):
    query = """MATCH (n:Company) where n.ticker = \"""" + ticker + """\" CALL apoc.path.subgraphAll(n, {labelFilter: "-Fact", maxLevel:2}) YIELD nodes, relationships RETURN nodes, relationships"""
    return session.run(query)
```

```python
for t in tickers:
    print(t)
    data = new_query(t).data()
    with open(f"all_subgraphs/subgraph_two_hop_{t}.json", "w") as f:
        json.dump(data, f)
```

```python
def get_all_related_companies(driver, c):
    records, _, _ = driver.execute_query(
        f"""MATCH (n:Company)-[]-()-[]-(m:Company) where n.ticker = "{c}" RETURN m""",
        database_="neo4j", routing_=RoutingControl.READ
    )
    return records
```

```python
for t in tickers:
    print(t)
    with GraphDatabase.driver(URI, auth=AUTH) as driver:
        rec = get_all_related_companies(driver, t)
#     print(rec)
    data = []
    for dd in rec:
        node = dd["m"]
        temp = {
            "element_id": node.element_id,
            "labels": list(node.labels),
            "properties": dict(node.items())   
        }
        data.append(temp)
    with open(f"related_companies/related_companies_{t}.json", "w") as f:
        json.dump(data, f)
```
